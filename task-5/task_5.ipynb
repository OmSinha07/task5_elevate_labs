# Task 5: Exploratory Data Analysis (EDA) on Titanic Dataset

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("train.csv")
```

```python
# Set plot style
sns.set(style="whitegrid")
```

```python
# A. Basic Exploration
print("Basic Info:")
print(df.info())
print("\nDescriptive Statistics:")
print(df.describe(include="all"))

# Value counts for categorical variables
print("\nSex distribution:")
print(df["Sex"].value_counts())
print("\nEmbarked distribution:")
print(df["Embarked"].value_counts())
print("\nPclass distribution:")
print(df["Pclass"].value_counts())
```

```python
# B. Univariate Analysis
# Histograms for numerical features
df.hist(bins=20, figsize=(14, 10), edgecolor='black')
plt.suptitle("Histograms of Numerical Features", fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# C. Boxplots
# Age vs Survived
sns.boxplot(x="Survived", y="Age", data=df)
plt.title("Age Distribution by Survival")
plt.show()

# Fare vs Pclass
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
# D. Pairplot
# Select numeric features
selected_features = ["Survived", "Pclass", "Age", "SibSp", "Parch", "Fare"]
sns.pairplot(df[selected_features], hue="Survived")
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
# Only select numeric columns for correlation
numeric_df = df.select_dtypes(include=['number'])

# Plot the correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
